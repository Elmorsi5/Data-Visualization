```python
df_mean=df.iloc[:,:11]
df_mean.drop(['perimeter_mean','area_mean','concave points_mean'],axis=1,inplace=True)
```

```python
def plott(co):
    plt.subplot(1,2,1)
    sns.distplot(df_mean[co][df_mean['diagnosis']=='M']).set_title('malignant')
    plt.subplot(1,2,2)
    sns.distplot(df_mean[co][df_mean['diagnosis']=='B']).set_title('benign')
    plt.show()
```
