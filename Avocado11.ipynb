# cleaning_arranging Data

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={'4046':'small ','4225':'large','4770':'extra large '},inplace=True)
```

```python
df.head()
```

```python
df[['Year','Month','Day']]=df['Date'].str.split('-',expand=True)
```

```python
df.head()
```

```python
df['Year']=pd.to_numeric(df['Year'])
df['Month']=pd.to_numeric(df['Month'])
df['Day']=pd.to_numeric(df['Day'])
```

# as we see there is no more need for Data and Year-It is Repeated- so we are going to remove it.

```python
df.drop(['Date','year'],axis=1,inplace=True)
```

```python
df.head()
```

# Treating With Miss Values

```python
df.isna().sum()
```

as we see"XLarge Bags","extra large" have more than 30% misses value, so we remove them..

```python
df.drop(['XLarge Bags','extra large '],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.info()
```

we are going to fill miss value with the median..

```python
df_median=df.median()
```

```python
df.fillna(df_median,inplace=True)
```
